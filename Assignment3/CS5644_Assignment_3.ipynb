```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn import model_selection
from sklearn.metrics import r2_score
from sklearn import neighbors
import collections
```

# Q1

```python
df_hour = pd.read_csv('./bike+sharing+dataset/hour.csv')
df_hour.head()
```

```python
df_day = pd.read_csv('./bike+sharing+dataset/day.csv')
df_day.head()
```

```python
df_hour_drop = df_hour.drop(['instant' , 'dteday','casual' , 'registered' , 'cnt'], axis=1)
df_hour_drop.head()
```

```python
df_day_drop = df_day.drop(['instant' , 'dteday' , 'casual' , 'registered' , 'cnt'], axis=1)
df_day_drop.head()
```

```python
print('hour')
models = [LinearRegression(),KNeighborsRegressor()]
predicts = ['casual' , 'registered' , 'cnt']
feature = list(df_hour_drop)
X=df_hour_drop[feature]
      
def train(model):
    y=df_hour[predict]
    kfold = model_selection.KFold(n_splits=5)
    r2 = model_selection.cross_val_score(model, X, y, cv=kfold, scoring='r2')
    neg_mean_squared_error = model_selection.cross_val_score(model, X, y, cv=kfold, scoring='neg_mean_squared_error')
    r2_mean = r2.mean()
    mse_mean=neg_mean_squared_error.mean()
    print('Model:',model)
    print('R2:', abs(r2_mean))
    print('MSE:', abs(mse_mean))
for model in models:
      for predict in predicts:
        print()
        print('Predicts:',predict)
        train(model)
```

```python
print('day')
models = [LinearRegression(),KNeighborsRegressor()]
predicts = ['casual' , 'registered' , 'cnt']
feature = list(df_day_drop)
X=df_day_drop[feature]
      
def train(model):
    y=df_day[predict]
    kfold = model_selection.KFold(n_splits=5)
    r2 = model_selection.cross_val_score(model, X, y, cv=kfold, scoring='r2')
    neg_mean_squared_error = model_selection.cross_val_score(model, X, y, cv=kfold, scoring='neg_mean_squared_error')
    r2_mean = r2.mean()
    mse_mean=neg_mean_squared_error.mean()
    print('Model:',model)
    print('R2:', abs(r2_mean))
    print('MSE:', abs(mse_mean))
for model in models:
      for predict in predicts:
        print()
        print('Predicts:',predict)
        train(model)
```

```python
df_seeds = pd.read_table('./seeds/seeds_dataset.txt', header = None, sep = "\\s+")
df_seeds.head()
```

```python
X_matrix=df_seeds.iloc[:,0:7]
X_matrix.head()
```

```python
y=df_seeds.iloc[:,7]
y
```

```python
knearest_clf = neighbors.KNeighborsClassifier(n_neighbors=3, weights='uniform') 
knearest_clf.fit(X_matrix, y)
```

```python
seeds_preds=knearest_clf.predict(X_matrix)
print("predicted:", seeds_preds)
print("truth", y)
```

```python
print("Clusters predict:")
print(collections.Counter(seeds_preds))
print("Truth:")
print(collections.Counter(y))
```
